# Linear Regression

Linear regression is a statistical method used for modeling the relationship between a dependent variable and 
one or more independent variables. The basic idea is to find the best-fitting linear equation that describes 
the relationship between the variables. In simple linear regression, there is only one independent variable, 
while in multiple linear regression, there are two or more.

The equation for a simple linear regression is often represented as:

y=mx+b

Where:
y is the dependent variable (the one you are trying to predict),
x is the independent variable (the one used to make predictions),
m is the slope of the line (the change in 
y with respect to a change in x),
b is the y-intercept (the value of y when x is 0).

Linear regression is widely used in various fields, including statistics, economics, biology, and machine learning. 
It is a fundamental tool for understanding and predicting relationships between variables.

```python
x = df['Weight']
y = df['Price']
```

Here, I assigned the 'Weight' column to the variable x and the 'Price' column to the variable y. These variables are often used as coordinates for plotting.

```python
mean_x = np.mean(x)
mean_y = np.mean(y)
print("Mean Value of x:",mean_x)
print("Mean Value of y:",mean_y)
```


Here, I calculating the mean of the 'Weight' and 'Price' variables using NumPy. The np.mean() function in NumPy calculates the arithmetic mean (average) of the values in an array or a pandas Series.

```python
dev_x = x - mean_x
dev_y = y - mean_y
```

Here, I calculating the deviations of each data point from the mean for both the 'Weight' and 'Price' variables. These deviations are often computed to analyze the spread or variability of the data.

```python
coefficient = np.sum(dev_x * dev_y) / np.sum(dev_x**2) 
coefficient
```

Here, I calculating the coefficient of correlation between the 'Weight' and 'Price' variables. 

```python
intercept = mean_y - (coefficient * mean_x)
intercept
```

The intercept represents the value of the dependent variable ('Price' in this case) when the independent variable ('Weight') is 0. In the context of linear regression, it's the y-intercept of the regression line.

```python
pred_y = (coefficient * 6) + intercept
print("Predicted Price for Weight 6 is:",pred_y)
```

Here, I use the linear regression equation to predict the 'Price' (y) for a given 'Weight' (x). 
The equation for a simple linear regression is:

pred_y=coefficient × Weight + intercept

Here I use specific 'Weight' value (in this case, 6)
